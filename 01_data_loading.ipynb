Feature Engineering

Feature 1 - utilization ratio

```python
df_final['utilization_ratio'] = (df_final['nooftrans'] - df_final['othershoptranscnt']) / df_final['noofrcs']
print(df_final['utilization_ratio'].describe())
```

```python
zero_util = df_final[df_final['utilization_ratio'] == 0]
print(f"Shops with 0 utilization: {len(zero_util)}")
print(zero_util[['shopno', 'nooftrans', 'noofrcs', 'othershoptranscnt']].head())
```

Feature 2- Protability Rate

```python
df_final['portability_rate'] = df_final['othershoptranscnt'] / df_final['nooftrans']
print(df_final['portability_rate'].describe())
```

Feature 3 - Rice per card

```python
df_final['total_rice'] = df_final['riceafsc'] + df_final['ricefsc'] + df_final['riceaap']
df_final['rice_per_card'] = df_final['total_rice'] / df_final['nooftrans']
print(df_final['rice_per_card'].describe())
```

Feature 4- Volatility

```python
shop_volatility = df_final.groupby('shopno')['nooftrans'].std()
print(shop_volatility.head())
print(shop_volatility.describe())
```

```python
shop_volatility = shop_volatility.reset_index()
shop_volatility.columns = ['shopno', 'trans_volatility']
print(shop_volatility.head())
```

```python
df_final = pd.merge(df_final, shop_volatility, on='shopno', how='inner')
print(df_final.shape)
```

```python
df_final = df_final.drop(columns=['total_rice'], errors='ignore')
print(df_final.shape)
print(df_final.columns.tolist())
```

Aggregation
To collapse 29 months into 1 row

```python
shop_profiles = df_final.groupby('shopno').agg({
    'utilization_ratio': 'mean',
    'portability_rate': 'mean',
    'rice_per_card': 'mean',
    'trans_volatility': 'first',
    'latitude': 'first',
    'longitude': 'first',
    'distname': 'first',
    'distcode': 'first'
}).reset_index()

print(shop_profiles.shape)
print(shop_profiles.head())
```

Scaling

"StandardScaler brings ALL features to the same scale so that no single feature dominates the distance calculation in K-Means"

```python
from sklearn.preprocessing import StandardScaler

features = ['utilization_ratio', 'portability_rate', 
            'rice_per_card', 'trans_volatility']

scaler = StandardScaler()
scaled_features = scaler.fit_transform(shop_profiles[features])
print(scaled_features.shape)
```


elbow method

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(K_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.xticks(K_range)
plt.grid(True)
plt.show()
```

k-means cluster = 5

```python
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(scaled_features)

shop_profiles['cluster'] = kmeans.labels_
print(shop_profiles['cluster'].value_counts())
```

Cluster summary -- grouping each cluster by feature engineering

```python
cluster_summary = shop_profiles.groupby('cluster')[features].mean()
print(cluster_summary)
```

Validation with shillouette score
shillouette score = (distance to other cluster - distance to own cluster) / max of both clusters

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(scaled_features, kmeans.labels_)
print(f"Silhouette Score: {score}")
```

DBSCAN - Density-Based Spatial Clustering of Applications with Noise

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_features)

# Check results
import numpy as np
unique, counts = np.unique(dbscan_labels, return_counts=True)
print(dict(zip(unique, counts)))
```

```python
# Add DBSCAN labels
dbscan = DBSCAN(eps=0.5, min_samples=5)
shop_profiles['dbscan_cluster'] = dbscan.fit_predict(scaled_features)

# Investigate noise shops
noise_shops = shop_profiles[shop_profiles['dbscan_cluster'] == -1]
print(f"Total noise shops: {len(noise_shops)}")
print(noise_shops[features].describe())
```

```python
most_suspicious = noise_shops.nlargest(5, 'trans_volatility')
print(most_suspicious[['shopno', 'distname', 
                        'utilization_ratio', 
                        'portability_rate',
                        'trans_volatility']])
```
